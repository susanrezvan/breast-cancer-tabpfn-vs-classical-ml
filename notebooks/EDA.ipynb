```python
# Import required libraries for data loading and visualization
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectFromModel
from sklearn.pipeline import Pipeline
```

```python
#Load the dataset from CSV file
df = pd.read_csv("/kaggle/input/breast-cancer-wisconsin-data/data.csv")
df.head()
```

```python
#Check the shape of the dataset
df.shape
```

```python
# Counts how many samples belong to each class: B (benign), M (malignant).
y = df['diagnosis']
y.value_counts()
```

```python
#Visualize class distribution with a countplot
# Helps to see imbalance between benign and malignant samples.
sns.countplot(x = df['diagnosis'],label = 'count', palette={'B': 'orange', 'M': 'blue'})
plt.show()
```

```python
# Visualizes relationships between features and how they differ
sns.countplot(x = df['diagnosis'],label = 'count', palette={'B': 'orange', 'M': 'blue'})
plt.show()
```

```python
# Pairplot of selected features grouped by diagnosis
# Visualizes relationships between features and how they differ
sns.pairplot(df.iloc[:,1:8], hue = 'diagnosis')
plt.show()
```
